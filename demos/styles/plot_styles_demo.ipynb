# Define a Plot for Testing

```python
def style_testplot(palette, ls, markers, cmap):
    """
        - function defining a testplot to compare different styles with each other
    """
    fig = plt.figure(figsize=(16,16))
    fig.suptitle("Suptitle")
    ax1 = fig.add_subplot(321, title="ax1", xlabel="x", ylabel="y")
    ax2 = fig.add_subplot(322, title="ax2", xlabel="x", ylabel="y", zlabel="z", projection="3d")
    ax3 = fig.add_subplot(323, title="ax3", xlabel="x", ylabel="y")
    ax4 = fig.add_subplot(324, title="ax4", xlabel="x", ylabel="y")
    ax5 = fig.add_subplot(325, title="ax5", xlabel="x", ylabel="Counts")
    ax6 = fig.add_subplot(326, title="ax6", xlabel="x", ylabel="Counts")
    for idx in range(10): ax1.plot(np.linspace(0,1,5)+idx, label=f"Line {idx+1}")    

    xx, yy = np.meshgrid(np.linspace(1,5,10), np.linspace(1,5,100))
    surf = ax2.plot_surface(xx, yy, yy, cmap=cmap)
    ax2.minorticks_off()
    cbar = fig.colorbar(surf, ax=ax2)
    cbar.set_label("Cbar")
    cbar.ax.tick_params(which="both", direction="out")
    cbar.ax.minorticks_off()

    mesh = ax3.pcolormesh(np.random.randn(50,50))
    cbar = fig.colorbar(mesh, ax=ax3)
    cbar.set_label("Cbar")
    cbar.ax.tick_params(which="both", direction="out")
    cbar.ax.minorticks_off()

    for idx in range(7): sctr = ax4.scatter(*np.random.randn(20,2).T, label=f"Scatter {idx+1}", marker=markers[idx])
    sctr = ax4.scatter(*np.random.randn(20,2).T, c=np.random.rand(20), label=f"Scatter {idx+1}", marker=markers[idx], cmap="coolwarm")
    cbar = fig.colorbar(sctr, ax=ax4)
    cbar.set_label("Cbar")
    cbar.ax.tick_params(which="both", direction="out")
    cbar.ax.minorticks_off()

    a = np.random.randn(100)
    ax5.hist(a, histtype="step")
    ax6.hist(a, color="C2")

    ax1.legend(ncol=2, title="LEGTITLE", loc="upper right")
    ax4.legend()
    fig.tight_layout()
    plt.show()

    return
```
